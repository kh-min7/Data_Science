### 정확성

이상점 : 다른 값들과 너무 동떨어져 있는 데이터

* 이상점이 잘못된 데이터라면?

고치거나 제거

* 이상점이 제대로 된 데이터라면?

분석에 방해가 되면 제거, 의미 있는 정보라면 두기

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('data/beer.csv', index_col=0)
df.head()
```

```python
df.plot(kind='box', y='abv')
```

```python
df['abv'].describe()
```

```python
q1 = df['abv'].quantile(0.25)
q3 = df['abv'].quantile(0.75)
iqr = q3 - q1
```

#### 이상점 구하기

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
```

```python
df[condition]
```

#### 이상점 고치기

```python
df.loc[2250, 'abv'] = 0.055
```

```python
df.loc[2250]
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
df[condition]
```

```python
df[condition].index
```

```python
df.drop(df[condition].index, inplace=True)
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
df[condition]
```

```python
df.plot(kind='box', y='abv')
```
